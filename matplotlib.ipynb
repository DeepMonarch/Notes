# Pyplot functions()

```python
# pyplot functions()
x = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
y = [10, 20, 15, 25, 20]

plt.plot(x, y) # plt.plot(x,y, color='green', marker='o', linestyle='dashed', linewidth=2,
#markersize=12, marker='o', markerfacecolor='blue', markeredgewidth=2, markeredgecolor='red', label='Sales Data'    ki)

plt.title("Bakery sales over a week")
plt.xlabel("Days of the week")
plt.ylabel("Sales per day")
plt.grid(True)
plt.legend(["Sales Data"])
plt.xlim(0, 6)
plt.ylim(0, 30)
plt.xticks([1,2,3,4,5], ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'])
plt.yticks([0,10,20,30], ['0', '10', '20', '30'])
plt.show()
```

```python
months = [1,2,3,4]
sales = [2500, 3000, 2800, 3500]

plt.plot(months, sales, marker='o', color='green', linestyle='dashed', linewidth=2, markersize=8, label='Sales Data')
plt.title("Monthly Sales Data")
plt.xlabel("Months")
plt.ylabel("Sales in USD")
plt.grid(True)
plt.legend() #upper left by default to change location use plt.legend(loc='upper right') or lower right etc.

# axes limits
plt.xlim(1, 4)
plt.ylim(2400, 3500)

# custom ticks
plt.xticks([1,2,3,4], ['Jan', 'Feb', 'Mar', 'Apr'])
plt.yticks([2400, 2600, 2800, 3000], ['2.4k', '2.6k', '2.8k', '3.0k'])

plt.show()
```

```python
# plt.bar(x, height, color='color name', width=value, label='label name')

import matplotlib.pyplot as plt

months = ['Jan', 'Feb', 'Mar', 'Apr']
sales = [2500, 3000, 2800, 3500]

plt.barh(months, sales, color='blue', label='Sales Data')
plt.title("Monthly Sales Data")
plt.xlabel("Months")
plt.ylabel("Sales in USD")
plt.grid(False)
plt.legend()

plt.show()
```

```python
# plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
import matplotlib.pyplot as plt

labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [30, 25, 25, 20]
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
# autopct shows percentage value on pie chart
# % = percentage .1 = one decimal place f = float format and the %% is to show the % sign
plt.title("Fruit Distribution")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# data in intervals are showed in histogram
import matplotlib.pyplot as plt

# plt.hist(data, bins=number_of_bins, color='color name', edgecolor='edge color', alpha=value)

scores = [55, 67, 45, 70, 89, 90, 78, 65, 88, 92, 73, 84, 77, 69, 95, 81, 60, 72, 85, 91]
plt.hist(scores, bins=5, color='purple', edgecolor='black', alpha=0.5)
plt.title("Distribution of Test Scores")
plt.xlabel("Score Ranges")
plt.ylabel("Number of Students")
plt.show()
```

## Subplot functions

```python
# subplot functions
# line chart (sales), bar chart (expenses), pie chart (revenue)
import matplotlib.pyplot as plt

sales = [2500, 3000, 2800, 3500]
expenses = [2000, 2500, 2300, 3000]


plt.subplot(1, 2, 1)  # 1 row, 2 columns, first subplot
plt.plot(sales, expenses, marker='o', color='green', label='Sales')
plt.title("Sales vs Expenses")
plt.xlabel("Months")
plt.ylabel("Amount in USD")
plt.legend()
plt.grid(True)


plt.subplot(1, 2, 2)  # 1 row, 2 columns, second subplot
plt.bar(['Jan', 'Feb', 'Mar', 'Apr'], expenses, color='orange', label='Expenses')
plt.title("Monthly Expenses")
plt.xlabel("Months")
plt.ylabel("Amount in USD")
plt.legend()
plt.grid(True)
plt.tight_layout()  # Adjust subplots to fit into figure area
plt.show()
```

```python
sales = [2500, 3000, 2800, 3500]
expenses = [2000, 2500, 2300, 3000]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5)) # fig, (ax1, ax2) are axes objects, figsize sets the size of the entire figure, 1 row, 2 columns

""" 
can also be written as:
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0] for first subplot
ax[1] for second subplot 
"""
fig.suptitle("Company Financial Overview")  # Overall title for the figure

ax1.plot(sales, expenses, marker='o', color='green', label='Sales')
ax1.set_title("Sales vs Expenses")
ax1.set_xlabel("Months")
ax1.set_ylabel("Amount in USD")
ax1.legend()
ax1.grid(True)
ax2.bar(['Jan', 'Feb', 'Mar', 'Apr'], expenses, color='orange', label='Expenses')
ax2.set_title("Monthly Expenses")
ax2.set_xlabel("Months")
ax2.set_ylabel("Amount in USD")
ax2.legend()
ax2.grid(True)

plt.savefig("financial_overview.jpeg", dpi=300, bbox_inches='tight')  # dpi sets resolution, bbox_inches='tight' trims extra whitespace
plt.tight_layout()  # Adjust subplots to fit into figure area
plt.show()

# Saving Figures
# plt.savefig("financial_overview.png")
# How to save the figure created above
```
