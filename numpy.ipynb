### Indexing and Slicing 
### Fancy indexing
### Boolean masking

```python
# indexing
""" array[index] # for 1-D array
    array[row, column] # for 2-D array"""
import numpy as np

#1-D array indexing
ar_1dd = np.array([10, 20, 30, 40, 50])
print("Element at index 2 in ar_1d:", ar_1dd[2])  # 30

#2-D array indexing
ar_2dd = np.array([[10, 20, 30],
                    [40, 50, 60]])
print("Element at row 1, column 2 in ar_2d:", ar_2dd[-1, 2])  # 60

#Slicing [start:stop:step], stop is exclusive.
#1-D array slicing
print("Elements from index 1 to 3 in ar_1d:", ar_1dd[1:4])  # [20 30 40]

#2-D array slicing
print("Elements from row 0, columns 1 to 2 in ar_2d:", ar_2dd[0, 1:3])  # [20 30]

#Advanced Indexing
print(ar_1dd[1:4:2])  # select every second element from index 1 to 3
print(ar_1dd[::-1])  # reverse the array
print(ar_2dd[::2, ::2])  # select every second row and every second column
```

```python
# fancy indexing
indices = [0, 2, 4]
print("Fancy indexing on ar_1d:", ar_1dd[indices])  # [10 30 50]

# boolean masking (filtering data based on conditions)
mask = ar_1dd > 25
print("Boolean masking on ar_1d:", ar_1dd[mask])  # [30 40 50]
```

### Reshaping & Manipulating Arrays!

```python
import numpy as np
# Changing the shape or dimensions of an array without changing its data
# It does not create a copy of the data, just a new view with a different shape
ar = np.array([[1, 2, 3, 4],
                [5, 6, 7, 8]])
# Reshaping
reshaped_ar = ar.reshape((4, 2))  # new shape (4 rows, 2 columns)
print("Reshaped Array (4x2):")
print(reshaped_ar)
```

### Flattening Array
### ravel(), flatten()

```python
# multidimensional array to 1-D array
"""ravel() -> view
flatten() -> copy"""

import numpy as np
ar_nd = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
# Using ravel()
raveled_ar = ar_nd.ravel()
print("Raveled Array:")
print(raveled_ar)

# Using flatten()
flattened_ar = ar_nd.flatten()
print("Flattened Array:")
print(flattened_ar)
```

### Array Modification

```python
# add, remove, stack, split arrays
import numpy as np


# Adding arrays
ar1 = np.array([1, 2, 3])
ar2 = np.array([4, 5, 6])
ar_add = np.add(ar1, ar2)
print("Addition of arrays:")
print(ar_add)

""" np.insert(array, index, value, axis=None)
    axis = None for 1-D array 
    axis = 0 for rows in 2-D array
    axis = 1 for columns in 2-D array 
"""

# Inserting elements
ar_insert = np.insert(ar1, 1, 10)  # insert 10 at index 1
print("Array after insertion:")
print(ar_insert)

# inserting a new row in 2-D array
ar_2d = np.array([[1, 2, 3],
                   [4, 5, 6]])
ar_insert_2d = np.insert(ar_2d, 1, [10, 11, 12], axis=0)  # insert new row at index 1
print("2-D Array after row insertion:")
print(ar_insert_2d)
ar_insert_2d = np.insert(ar_2d, 1, [10, 11], axis=1)  # insert new column at index 1
print("2-D Array after column insertion:")
print(ar_insert_2d)

# Removing elements
ar_remove = np.delete(ar1, 1)  # remove element at index 1
print("Array after removal:")
print(ar_remove)

# Stacking arrays
ar_stack = np.vstack((ar2, ar1)) # vertical stack it stacks arrays row-wise and stacking means putting them on top of each other
print("Stacked arrays:")
print(ar_stack)

# Splitting arrays
""" np.arr.split() -> equally split
    np.arr.hsplit() -> horizontal split
    np.arr.vsplit() -> vertical split
"""
ar_split = np.split(ar1, 3)  # split into 3 parts
print("Split arrays:")
print(ar_split)

# appending elements
ar_append = np.append(ar1, [7, 8, 9], axis=0)
print("Array after appending elements:")
print(ar_append)

#concatenating arrays
ar_concat = np.concatenate((ar1, ar2), axis=0)
print("Array after concatenating:")
print(ar_concat)

# deleting elements, and rows/columns
ar_delete = np.delete(ar_2d, 0, axis=0)  # delete first row
print("2-D Array after deleting first row:")
print(ar_delete)
```

### Broadcasting and Vectorization

```python
prices = [100, 150, 200, 250, 300]
discount = 10  # 10% discount
final_prices = []

for price in prices:
    discounted_price = price - (price * discount / 100)
    final_prices.append(discounted_price)

print(final_prices)

# using broadcasting
import numpy as np
prices_array = np.array(prices)
final_prices_array = prices_array - (prices_array * discount / 100)
print(final_prices_array)
```

## How numpy handles arrays of different dimensions

```python
## How numpy handles arrays of different dimensions
# Matching dimensions
ar1 = np.array([[1, 2, 3],
                 [4, 5, 6]])
ar2 = np.array([[10, 20, 30],
                 [40, 50, 60]])

ar_sum = ar1 + ar2  # no broadcasting, shapes are compatible
print("Sum of ar1 and ar2 with matching dimensions:")
print(ar_sum)

# Expanding single elements
ar3 = np.array([[100],
                 [200]])
ar_combined = ar1 + ar3  # broadcasting happens here
print("Sum of ar1 and ar3 with broadcasting:")
print(ar_combined)

# Incompatible shapes
ar4 = np.array([[1, 2],
                 [3, 4],
                 [5, 6]])
# This will raise an error due to incompatible shapes
# ar_invalid = ar1 + ar4
```

```python
# 1d to 2d

a = np.array([1, 2, 3])
b = np.array([[1, 2, 3],
                 [4, 5, 6]])
c = a + b  # broadcasting happens here
print("Sum of 1-D array a and 2-D array b with broadcasting:")  
print(c)

ab = np.array([1, 2, 3])
cd = np.array([1,2])
# This will raise an error due to incompatible shapes
# result = ab + cd

# solving incompatible shapes by reshaping
cd_reshaped = cd.reshape((2, 1))  # reshape cd to (2, 1)
print("Reshaped cd:")
print(cd_reshaped)
result = ab + cd_reshaped  # broadcasting happens here
print("Sum of ab and reshaped cd with broadcasting:")
print(result)
```

# vectorization

```python
# vectorization is the process of converting iterative operations
# into array operations to improve performance and efficiency.
import numpy as np
# Without vectorization
numbers = np.array([1, 2, 3, 4, 5])
squared = np.array([x**2 for x in numbers])
print("Squared numbers without vectorization:")
print(squared)

# With vectorization [much faster]
squared_vectorized = numbers ** 2
print("Squared numbers with vectorization:")
print(squared_vectorized)
```

## Handling missing values

```python
# Handling missing values using numpy's built-in functions
import numpy as np
# isnan(), isfinite(), isinf()
"""
np.isnan() -> detect missing values (NaN)
np.nan_to_num() -> replace NaN with specified value
np.isfinite() -> detect finite values (not NaN or Inf)
np.isinf() -> detect infinite values
"""

data = np.array([1.0, 2.0, np.nan, 4.0, np.inf, -np.inf])

cleaned_data = np.nan_to_num(data, nan=0.0, posinf=1e10, neginf=-1e10)
print("Cleaned Data:")
print(cleaned_data)

# Detecting missing values
nan_mask = np.isnan(data)
print("NaN mask:")
print(nan_mask)
# Replacing missing values with zero
data_no_nan = np.nan_to_num(data, nan=0.0)
print("Data after replacing NaN with 0:")
print(data_no_nan)
# Detecting finite values
finite_mask = np.isfinite(data)
print("Finite mask:")
print(finite_mask)
# Detecting infinite values
inf_mask = np.isinf(data)
print("Infinite mask:")
print(inf_mask)
# Replacing infinite values with large finite numbers
data_no_inf = np.where(np.isinf(data), np.sign(data) * 1e10, data)
print("Data after replacing Inf with large finite numbers:")
print(data_no_inf)
```
