```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({
    'cgpa': [3.5, 3.0, 2.5, 3.8, 2.0],
    'resume_score': [8, 7, 6, 9, 5],
    'placed': [1, 1, 0, 1, 0]
})
```

```python
print(df.shape)
df.head()
```

```python
sns.scatterplot(x='cgpa', y='resume_score', hue='placed', data=df)
plt.title('CGPA vs Resume Score')
```

```python
# using scikit learn perceptron class
x = df.iloc[:,0:2]
y = df.iloc[:,-1]
```

```python
from sklearn.linear_model import Perceptron
model = Perceptron()
model.fit(x,y)
```

```python
model.coef_ #w1 = 1.9, w2 = 2.
```

```python
model.intercept_ #b = -3.
```

```python
%pip install mlxtend
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x.values, y.values, clf=model)
```
