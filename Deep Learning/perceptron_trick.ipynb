```python
from sklearn.datasets import make_classification
import numpy as np
X, y = make_classification(n_samples=100, n_features=2, n_classes=2, n_informative=1,
                           n_redundant=0, n_clusters_per_class=1, random_state=42, hypercube=False, class_sep=2)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.scatter(X[:,0], X[:,1], c=y, cmap='winter', s=100)
```

```python
def perceptron(X,y):
    
    x = np.insert(X, 0, 1, axis=1) # add bias term
    w = np.ones(X.shape[1]) # initialize weights
    lr = 0.1 # learning rate
    
    for i in range(1000):
        j = np.random.randint(0,100) # random sample 
        y_pred = step(np.dot(x[j], w))
        w += lr * (y[j] - y_pred) * x[j] # update weights
    return w[0], w[1:]
```

```python
def step(z):
    return 1 if z >= 0 else 0
```

```python
intercept_, coef_ = perceptron(X,y)
```
