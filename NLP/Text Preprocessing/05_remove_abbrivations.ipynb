### Removing short forms & abbreivations from chat:

#### converting slang.txt to dict

```python
# Assuming your file is named 'slang.txt'
chatwords = {}

with open('slang.txt', 'r') as f:
    # Split each line at '=' and strip whitespace
    chatwords = {line.split('=')[0]: line.split('=')[1].strip() 
                  for line in f if '=' in line}

print(chatwords['AFK']) # Output: Away From Keyboard
```

```python
# ex. rofl, lmao, imho, fyi
# A dictionary of chat words and their meanings
# github.com/rishabhverma17/sms_slang_translator/blob/master/slang.txt

def chat_conversion(text):
    """Convert chat language to normal language"""
    new_text = []
    for word in text.split():
        if word.upper() in chatwords:
            new_text.append(chatwords[word.upper()])
        else:
            new_text.append(word)
    return ' '.join(new_text)
    
```

```python
chat_conversion("I need to check something. AFK for a bit.")
```
