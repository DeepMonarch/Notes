### Using Porter Stemmer for english
##### for other language use- Snow Ball Stemmer

```python
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
print(ps.stem('happiness'))
```

```python
def stem_words(text):
    return ' '.join([ps.stem(word) for word in text.split()])

sample = "walk walked walking walks"
print(stem_words(sample))
```
