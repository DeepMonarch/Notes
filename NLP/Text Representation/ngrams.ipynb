### Bag of N-Grams:

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    'text': ['I love NLP', 'NLP is great', 'I enjoy learning NLP'],
    'output': [1, 1, 1]
})
```

```python
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(ngram_range=(2,2)) # bi-gram # if (1,2) both unigrams and bigrams
```

```python
bon = cv.fit_transform(df['text'])
```

```python
print(cv.vocabulary_)
```
