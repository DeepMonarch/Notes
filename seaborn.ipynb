## Histogram

```python
# Histogram in Seaborn

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset("penguins")
# bins fix range ke liye
# kde kolor density estimation ratio line chart bana deta hai
# rug for colnol density maybe it shows lines in bottom 
# log_scale=True log ke scale me dekhne ke liye

sns.displot(df['bill_length_mm'], bins = [40, 45, 55, 60], kde=True, rug=True, color='r')

plt.show()
```

## Barplot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset("penguins")

order_1 = ["Dream", "Torgersen", "Biscoe"]
sns.set(style='darkgrid')
sns.barplot(x = df.island,
            y = df.bill_length_mm,
            data=df,
            hue='sex',
            order=order_1,
            hue_order=['Female', 
            "Male"],
            errorbar=('ci',100),
            n_boot=2, orient="v",
            palette='dark:g',
            saturation=0.3,
            err_kws={'color': 'g', 'linewidth': 12},
            capsize=0.5,
            dodge=False,
            alpha = 0.5
            ) # or use data = df necessary for hue
# hue se penguins me male kitne hai or female kitne hai wo dekha # dodge to reverse it
# order for hue # ci for black line on them
# for h horizontal orientation both the values should be numeric
# saturatio 0 to 1
# errcolor and errwidth for the black line (error)

plt.show()
```

## Scatter Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('penguins').head(20)

m = { 'Male':'o', 'Female':'*'}
sns.scatterplot(x = df.bill_length_mm,
                 y= df.bill_depth_mm,
                   data = df, hue='sex',
                     style = 'sex',
                       size = 'sex', sizes = (100, 40),
                        palette='terrain',
                          alpha = 0.5,
                           markers=m)
plt.show()
```

## Heatmap

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
arr = np.linspace(1,10,20).reshape(4,5)
arr
```

```python
sns.heatmap(arr)
```

```python
data = sns.load_dataset('anagrams').head(10)
x = data.drop(columns=['attnr'], axis=1)
```

```python
sns.heatmap(x, vmin=0, vmax=20,
            cmap='gist_heat',
            annot = True,
            )
```

```python
var = np.linspace(1,10,10).reshape(2,5)
var
```

```python
a = np.array([['a0','a1','a2','a3','a4'], ['b0','b1','b2','b3','b4']])
y = {'fontsize':20, 'color':'r'}
v = sns.heatmap(var,
            cmap='gist_heat',
            annot = a,
            fmt= 's', # s = string type data
            annot_kws=y,
            linewidth=10,
            linecolor='y',
            cbar=False ,# to remove the color bar 
            xticklabels=False, # ytickslabels false se y ke label hat jayege

            )

v.set(xlabel = 'python', ylabel ='conda')
sns.set_theme(font_scale=1)
plt.show()
```
